```python
label_encoder = LabelEncoder()
```

```python
for column in df.columns:
    if df[column].dtype not in ['int64', 'float64']:
        df.loc[:, column] = label_encoder.fit_transform(df[column])
```

```python
scaler = StandardScaler()
```
